```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1 = pd.read_csv("tmdb_5000_movies.csv")
df2 = pd.read_csv("tmdb_5000_credits.csv")
movies= df1.merge(df2,on="title")
# movies.shape
movies.head(2)
# working with : # genres , id , keywords , title ,overview , cast , crew
movies = movies[["id","genres","keywords","title","overview","cast","crew"]]
movies.head(5)
```

```python
movies.isnull().sum()
```

```python
import ast
movies.dropna(inplace=True)
# movies.duplicated()
# movies.isnull().sum()
movies.loc[0,"genres"]
def get_name_from_dict(obj):
    genre_list = []
    for i in ast.literal_eval(obj):
        genre_list.append(i["name"])
    return genre_list
movies["genres"]=movies["genres"].apply(get_name_from_dict)
```

```python
# movies.head()
movies["keywords"] = movies["keywords"].apply(get_name_from_dict)
```

```python
def get_name_from_cast(obj):
    genre_list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            genre_list.append(i["name"])
            counter +=1
        else:
            break
    return genre_list
movies["cast"] = movies["cast"].apply(get_name_from_cast)
movies["cast"].head(1)
# movies.head()
```

```python
def fetching_director_from_crew(obj):
    genre_list = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            genre_list.append(i["name"])
            break
    return genre_list
movies["crew"] = movies["crew"].apply(fetching_director_from_crew)
movies["crew"].head(1)
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split()) 
```

```python
# movies.head()
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ","") for i in x])
movies.head()
```

```python
movies["tag"] = movies["overview"] + movies["genres"]  + movies["keywords"] +  movies["cast"] +  movies["crew"]
new_df = movies[["id","title","tag"]]
```

```python
new_df["tag"] = new_df["tag"].apply(lambda x:" ".join(x))
```

```python
# new_df["tag"][0]
new_df = new_df.rename(columns={"tag":"tags"})
new_df["tags"] = new_df["tags"].str.lower() 
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))
    return " ".join(x)
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
# new_df["tags"][2]
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[1245]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity = similarity.astype(np.float32)
```

```python
# similarity
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movies):
    movie_index = new_df[new_df["title"] == movies].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
recommend("Pirates of the Caribbean: At World's End")
```

```python
recommend("The Dark Knight Rises")
```
